#Basic Image Processing Operations

We will be demonstrating some basic image processing operations using OpenCV and Matplotlib.

```python
import cv2
from matplotlib import pyplot as plt 
```

##Uploading and displaying an image
We read an image using OpenCV.

```python
img_bgr = cv2.imread('/content/drive/MyDrive/images/small_emilia.jpg')

plt.imshow(img_bgr)
plt.show()
```

OpenCV uploads images as Blue, Green, Red (BGR) instead of the traditional order, RGB. We need to get the arrays in the right order so that we can correctly visualize the image.

```python
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.show()
```

##Image info
We can get some info about our image.

```python
h, w, c = img_rgb.shape
print("Height: {:d}, width: {:d}, number of channels: {:d}".format(h, w, c))
```

```python
# To get one pixel: img_rgb[chosen_h, chosen_w]

dark_pixel = img_rgb[100, 200]
print("Dark pixel: ", dark_pixel)

clear_pixel = img_rgb[500, 100]
print("Clear pixel: ", clear_pixel)
```

##Greyscale
We convert our colour image into a greyscale image.

```python
img_grey = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)

plt.imshow(img_grey, cmap='gray')
plt.show()
```

```python
dark_pixel = img_grey[100, 200]
print("Dark pixel: ", dark_pixel)

clear_pixel = img_grey[500, 100]
print("Clear pixel: ", clear_pixel)
```

##Histogram
Computing the histogram of the RGB image

```python
color = ('r', 'g','b')
for i, col in enumerate(color):
    histr = cv2.calcHist([img_rgb], [i], None, [256], [0, 256])
    plt.plot(histr, color = col)
    plt.xlim([0, 256])
plt.show()
```
